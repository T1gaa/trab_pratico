```python
import matplotlib.pyplot as plt
import numpy as np

# Example data
variables = ['lr', 'svc', 'bagg_dt', 'rf','gbc','xgboost','stacking','softV']  # Variables
categories = ['local', 'public', 'private']   # Categories
values = [
    [0.354, 0.390, 0.279],  # Values for lr
    [0.347, 0.287, 0.351],  # Values for svc
    [0.352, 0.342, 0.348],  # Values for bagg_dt
    [0.326, 0.350, 0.371],  # Values for rf
    [0.339, 0.314, 0.359],  # Values for gbc
    [0.346, 0.357, 0.265],  # Values for xgboost
    [0.327, 0.306, 0.376],  # Values for stacking
    [0, 0.275, 0.340],  # Values for softVoting

]

# Define colors for each category
colors = ['blue', 'green', 'orange']

# Number of variables
x = np.arange(len(variables))  # x positions for the variables
width = 0.2  # Width of each bar

# Create the plot
fig, ax = plt.subplots(figsize=(8, 5))

# Plot bars for each category
for i, category in enumerate(categories):
    ax.bar(x + i * width, [value[i] for value in values], width, label=category, color=colors[i])

# Add labels and title
ax.set_xlabel('Models')
ax.set_ylabel('Macro f1')
ax.set_title('Performance macro f1')
ax.set_xticks(x + width)  # Center the ticks with the groups of bars
ax.set_xticklabels(variables)
ax.legend(title='Test')

# Show the plot
plt.tight_layout()
plt.show()
```
